```python
%matplotlib inline
import os
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_auc_score
from model import FLAIRModel
from torch.utils.data import Dataset, DataLoader, Subset, WeightedRandomSampler
from torchvision import transforms
from PIL import Image
import pandas as pd
import ast
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import roc_curve, auc, f1_score
from sklearn.preprocessing import label_binarize
import time
import logging
from collections import Counter

# =============================
# LOGGING SETUP
# =============================
logging.basicConfig(filename='image_errors.log', level=logging.ERROR,
                    format='%(asctime)s - %(levelname)s - %(message)s')

# =============================
# FOCAL LOSS
# =============================
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=2.0, reduction='mean'):
        super(FocalLoss, self).__init__()
        self.alpha = alpha
        self.gamma = gamma
        self.reduction = reduction

    def forward(self, inputs, targets):
        CE_loss = F.cross_entropy(inputs, targets, weight=self.alpha, reduction='none')
        pt = torch.exp(-CE_loss)
        focal_loss = ((1 - pt) ** self.gamma) * CE_loss
        if self.reduction == 'mean':
            return focal_loss.mean()
        elif self.reduction == 'sum':
            return focal_loss.sum()
        else:
            return focal_loss

# =============================
# CBAM ATTENTION MODULE
# =============================
class CBAM(nn.Module):
    def __init__(self, channels, reduction=16):
        super(CBAM, self).__init__()
        self.channel_gate = nn.Sequential(
            nn.AdaptiveAvgPool2d(1),
            nn.Conv2d(channels, channels // reduction, 1),
            nn.ReLU(),
            nn.Conv2d(channels // reduction, channels, 1),
            nn.Sigmoid()
        )
        self.spatial_gate = nn.Sequential(
            nn.Conv2d(channels, 1, 7, padding=3),  # 7x7 Ïª§ÎÑêÏùÑ 3x3ÏúºÎ°ú Î≥ÄÍ≤Ω, padding=1Î°ú Ï°∞Ï†ï
            nn.Sigmoid()
        )

    def forward(self, x):
        channel_attention = self.channel_gate(x)
        x = x * channel_attention
        spatial_attention = self.spatial_gate(x)
        x = x * spatial_attention
        return x

# =============================
# FCN-CLIP HYBRID MODEL
# =============================
class FCN_CLIP(nn.Module):
    def __init__(self, clip_model, num_classes=5, dropout_rate=0.3):
        super(FCN_CLIP, self).__init__()
        self.clip_model = clip_model
        self.vision_model = clip_model.vision_model
        self.feature_map_layer3 = None
        self.feature_map_layer2 = None
        if hasattr(self.vision_model.model, 'layer3'):
            self.vision_model.model.layer3.register_forward_hook(self.get_layer3_features)
        if hasattr(self.vision_model.model, 'layer2'):
            self.vision_model.model.layer2.register_forward_hook(self.get_layer2_features)

        self.use_fusion = False
        self.fusion = nn.Conv2d(1024 + 512, 1024, 1)

        self.decoder = nn.Sequential(
            nn.Conv2d(1024, 512, 3, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU(),
            nn.Dropout2d(dropout_rate),
            nn.Upsample(scale_factor=2, mode='bilinear'),
            nn.Conv2d(512, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU()
        )

        self.attention = CBAM(128)

        self.classifier = nn.Sequential(
            nn.AdaptiveAvgPool2d((1, 1)),
            nn.Flatten(),
            nn.Linear(128, num_classes)
        )

    def get_layer3_features(self, module, input, output):
        self.feature_map_layer3 = output

    def get_layer2_features(self, module, input, output):
        self.feature_map_layer2 = output

    def forward(self, x):
        with torch.no_grad():
            clip_features = self.vision_model(x)
        if self.feature_map_layer3 is not None:
            if self.use_fusion and self.feature_map_layer2 is not None:
                layer3 = F.interpolate(self.feature_map_layer3, size=self.feature_map_layer2.shape[2:], mode='bilinear')
                features = torch.cat([layer3, self.feature_map_layer2], dim=1)
                features = self.fusion(features)
            else:
                features = self.feature_map_layer3
        else:
            features = clip_features.view(clip_features.size(0), 1024, 1, 1)
            features = F.interpolate(features, scale_factor=32, mode='nearest')
        x = self.decoder(features)
        attention_map = self.attention(x)
        x = x * attention_map
        outputs = self.classifier(x)
        return outputs, clip_features

    def compute_contrastive_loss(self, image_features, text_embeddings, labels, temperature=0.1):
        image_features = image_features / image_features.norm(dim=-1, keepdim=True)
        text_embeddings = text_embeddings / text_embeddings.norm(dim=-1, keepdim=True)
        logits = (image_features @ text_embeddings.T) / temperature
        loss = F.cross_entropy(logits, labels)
        return loss * 0.01

# =============================
# CONFIG
# =============================
csv_files = ["./02_MESSIDOR.csv", "./train.csv"]
image_dirs = ["./IMAGES", "./train_images"]
combined_csv = "./combined_dataset.csv"

categories = [
    "no diabetic retinopathy",
    "mild diabetic retinopathy",
    "moderate diabetic retinopathy",
    "severe diabetic retinopathy",
    "proliferative diabetic retinopathy"
]
targets_dict = {cat: i for i, cat in enumerate(categories)}
device = 'cuda' if torch.cuda.is_available() else 'cpu'
temperature = 0.2
alpha = 0.7
batch_size = 32
test_batch_size = 16
EPOCHS = 20
patience = 7  # Increased for more convergence
num_classes = len(categories)

# =============================
# CSV PREPROCESSING
# =============================
def preprocess_csv(csv_files, root_dirs, output_csv):
    start_time = time.time()
    combined_data = []
    for csv_file, root_dir in zip(csv_files, root_dirs):
        print(f"üîç Reading CSV: {csv_file} with root dir: {root_dir}")
        df = pd.read_csv(csv_file)
        if 'categories' in df.columns:
            paths = df['image'].tolist()
            labels = [targets_dict[ast.literal_eval(cat)[0]] for cat in df['categories'].tolist()]
        else:
            paths = [f"{p}.png" for p in df['id_code'].tolist()]
            labels = df['diagnosis'].tolist()
        for path, label in zip(paths, labels):
            full_path = os.path.join(root_dir, path)
            if not os.path.exists(full_path):
                print(f"‚ùå File does not exist: {full_path}")
                logging.error(f"File does not exist: {full_path}")
                continue
            combined_data.append({'image': full_path, 'label': label})
    pd.DataFrame(combined_data).to_csv(output_csv, index=False)
    print(f"‚úÖ Combined CSV saved to {output_csv}, time: {time.time() - start_time:.2f}s")

if not os.path.exists(combined_csv):
    preprocess_csv(csv_files, image_dirs, combined_csv)

# =============================
# CUSTOM DATASET
# =============================
class CombinedRetinalDataset(Dataset):
    def __init__(self, csv_file, transform=None, minority_classes=[1, 3, 4]):
        self.transform = transform
        self.minority_classes = minority_classes
        self.minority_transform = transforms.Compose([
            transforms.RandomHorizontalFlip(p=0.6),
            transforms.RandomRotation(35),
            transforms.RandomResizedCrop(512, scale=(0.65, 1.0)),
            transforms.ColorJitter(brightness=0.25, contrast=0.25, saturation=0.25, hue=0.12),
            transforms.RandomAffine(degrees=18, translate=(0.12, 0.12), scale=(0.75, 1.25)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            transforms.RandomErasing(p=0.4, scale=(0.02, 0.12))
        ])
        self.image_paths = []
        self.labels = []
        start_time = time.time()
        print(f"üîç Reading CSV: {csv_file}")
        df = pd.read_csv(csv_file)
        self.image_paths = df['image'].tolist()
        self.labels = df['label'].tolist()
        print(f"‚úÖ Loaded {len(self.image_paths)} images, time: {time.time() - start_time:.2f}s")

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        path = self.image_paths[idx]
        label = self.labels[idx]
        if not os.path.exists(path):
            print(f"‚ùå File does not exist [{idx}]: {path}")
            logging.error(f"File does not exist [{idx}]: {path}")
            image = Image.new('RGB', (512, 512), color='black')
        else:
            try:
                with Image.open(path) as img:
                    image = img.convert('RGB')
            except Exception as e:
                print(f"‚ùå Failed to load image [{idx}]: {path}, Error: {e}")
                logging.error(f"Failed to load image [{idx}]: {path}, Error: {e}")
                image = Image.new('RGB', (512, 512), color='black')
        if self.transform and label in self.minority_classes:
            try:
                image = self.minority_transform(image)
            except Exception as e:
                print(f"‚ùå Transform failed on {path}: {e}")
                logging.error(f"Transform failed on {path}: {e}")
                image = torch.zeros(3, 512, 512)
        elif self.transform:
            try:
                image = self.transform(image)
            except Exception as e:
                print(f"‚ùå Transform failed on {path}: {e}")
                logging.error(f"Transform failed on {path}: {e}")
                image = torch.zeros(3, 512, 512)
        return {"image": image, "label": label}

    def get_labels(self):
        return self.labels

# =============================
# OVERSAMPLING
# =============================
def oversample_dataset(dataset, target_counts):
    if isinstance(dataset, Subset):
        labels = [dataset.dataset.labels[i] for i in dataset.indices]
        original_indices = dataset.indices
    else:
        labels = dataset.labels
        original_indices = list(range(len(dataset)))

    counts = Counter(labels)
    print(f"Original class distribution: {counts}")

    indices = []
    for label in range(num_classes):
        label_indices = [i for i, l in enumerate(labels) if l == label]
        if not label_indices:
            print(f"‚ö†Ô∏è Class {label} has no samples, skipping")
            continue
        label_original_indices = [original_indices[i] for i in label_indices]
        current_count = counts.get(label, 0)
        target_count = target_counts.get(label, 0)
        repeat = (target_count // current_count) + 1 if current_count > 0 else 0
        print(f"[DEBUG] Class {label}: Current {current_count}, Target {target_count}, Repeat {repeat}")
        for _ in range(target_count):
            indices.append(label_original_indices[_ % len(label_original_indices)])

    oversampled_indices = indices
    new_labels = [dataset.dataset.labels[i] for i in oversampled_indices]
    new_counts = Counter(new_labels)
    print(f"Oversampled class distribution: {new_counts}")

    for label in range(num_classes):
        expected = target_counts.get(label, 0)
        actual = new_counts.get(label, 0)
        if actual != expected:
            print(f"‚ö†Ô∏è Class {label}: Expected {expected}, Got {actual}")

    return Subset(dataset.dataset, oversampled_indices)

# =============================
# DATA AUGMENTATION
# =============================
train_transform = transforms.Compose([
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomVerticalFlip(p=0.5),
    transforms.RandomRotation(30),
    transforms.RandomResizedCrop(512, scale=(0.7, 1.0)),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),
    transforms.RandomAffine(degrees=15, translate=(0.1, 0.1), scale=(0.8, 1.2)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    transforms.RandomErasing(p=0.3, scale=(0.02, 0.1))
])

val_test_transform = transforms.Compose([
    transforms.Resize((512, 512)),
    transforms.RandomHorizontalFlip(p=0.1),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# =============================
# DATASET AND DATALOADER
# =============================
def compute_class_weights(dataset):
    start_time = time.time()
    if isinstance(dataset, Subset):
        labels = [dataset.dataset.labels[i] for i in dataset.indices]
    else:
        labels = dataset.labels
    counts = np.bincount(labels, minlength=num_classes)
    weights = 1.0 / (counts + 1e-6) ** 0.3
    # Moderately boost minority classes
    weights[1] *= 1.2  # Mild
    weights[3] *= 1.5  # Severe
    weights[4] *= 1.5  # Proliferative
    norm_weights = weights / weights.sum()
    sample_weights = torch.tensor([weights[label] for label in labels])
    print(f"[DEBUG] compute_class_weights time: {time.time() - start_time:.2f}s")
    print("Class distribution:", counts)
    return torch.tensor(norm_weights, dtype=torch.float), sample_weights

# Create dataset
full_dataset = CombinedRetinalDataset(combined_csv, transform=None)
num_total = len(full_dataset)
num_train = int(0.7 * num_total)
num_val = int(0.15 * num_total)
num_test = num_total - num_train - num_val

train_idx, val_idx, test_idx = torch.utils.data.random_split(
    range(num_total), [num_train, num_val, num_test], generator=torch.Generator().manual_seed(42)
)

train_dataset = Subset(full_dataset, train_idx)
val_dataset = Subset(full_dataset, val_idx)
test_dataset = Subset(full_dataset, test_idx)

print(f"‚úÖ Loaded {len(train_dataset) + len(val_dataset) + len(test_dataset)} images from datasets")

# Oversampling
target_counts = {0: 445, 1: 200, 2: 200, 3: 180, 4: 180}
train_dataset = oversample_dataset(train_dataset, target_counts)

# Apply transforms
train_dataset.dataset.transform = train_transform
val_dataset.dataset.transform = val_test_transform
test_dataset.dataset.transform = val_test_transform

# Compute weights and apply WeightedRandomSampler
class_weights, sample_weights = compute_class_weights(train_dataset)
sampler = WeightedRandomSampler(weights=sample_weights, num_samples=len(sample_weights), replacement=True)

print(f"[DEBUG] Sample weight length: {len(sample_weights)}, Train set length: {len(train_dataset)}")
print(f"[DEBUG] First 10 weights: {sample_weights[:10]}")

# DataLoaders
loaders = {
    "train": DataLoader(train_dataset, batch_size=batch_size, sampler=sampler, num_workers=4, pin_memory=True),
    "val": DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=4, pin_memory=True),
    "test": DataLoader(test_dataset, batch_size=test_batch_size, shuffle=False, num_workers=4, pin_memory=True),
}

# =============================
# MODEL INITIALIZATION
# =============================
clip_model = FLAIRModel(
    vision_type='resnet_v1',
    from_checkpoint=True,
    weights_path=None,
    image_size=512,
    caption="Fundus image showing [CLS]",
    projection=True,
    norm_features=True
).to(device)

model = FCN_CLIP(clip_model, num_classes=len(categories)).to(device)

if os.path.exists("best_fcn_clip_model.pth"):
    checkpoint = torch.load("best_fcn_clip_model.pth")
    model.load_state_dict(checkpoint, strict=False)
    print("‚úÖ Loaded best_fcn_clip_model.pth with strict=False")
    model.use_fusion = False
else:
    print("No checkpoint found, enabling layer2 fusion")
    model.use_fusion = True

nn.init.kaiming_normal_(model.attention.spatial_gate[0].weight, mode='fan_out', nonlinearity='relu')
if model.attention.spatial_gate[0].bias is not None:
    nn.init.constant_(model.attention.spatial_gate[0].bias, 0)

text_embed_dict, text_embeds = clip_model.compute_text_embeddings(categories, domain_knowledge=True)
text_embeds = text_embeds.to(device)

criterion = FocalLoss(alpha=class_weights.to(device), gamma=2.0)
optimizer = torch.optim.AdamW(model.parameters(), lr=2e-4, weight_decay=1e-3)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=2)

# =============================
# TRAINING FUNCTIONS
# =============================
def train_hybrid_epoch(model, dataloader, text_embeddings, optimizer, criterion, temperature=0.1, alpha=0.7):
    model.train()
    total_loss = 0
    total_ce_loss = 0
    total_contrastive_loss = 0
    all_preds, all_labels = [], []
    print(f"[DEBUG] Starting training epoch, dataloader size: {len(dataloader)}")
    for batch in tqdm(dataloader, desc="Training"):
        images = batch["image"].to(device)
        labels = batch["label"].to(device)
        optimizer.zero_grad()
        outputs, clip_features = model(images)
        ce_loss = criterion(outputs, labels)
        contrastive_loss = model.compute_contrastive_loss(clip_features, text_embeddings, labels, temperature)
        loss = alpha * ce_loss + (1 - alpha) * contrastive_loss
        loss.backward()
        total_norm = 0
        for p in model.parameters():
            if p.grad is not None:
                total_norm += p.grad.norm().item() ** 2
        total_norm = total_norm ** 0.5
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=2.0)
        optimizer.step()
        total_loss += loss.item()
        total_ce_loss += ce_loss.item()
        total_contrastive_loss += contrastive_loss.item()
        preds = torch.argmax(outputs, dim=1)
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

    avg_loss = total_loss / len(dataloader)
    avg_ce_loss = total_ce_loss / len(dataloader)
    avg_contrastive_loss = total_contrastive_loss / len(dataloader)
    acc = accuracy_score(all_labels, all_preds)
    print(f"[DEBUG] Avg CE Loss: {avg_ce_loss:.4f}, Avg Contrastive Loss: {avg_contrastive_loss:.4f}")
    return avg_loss, acc

def validate(model, dataloader, text_embeddings, criterion, temperature=0.1, alpha=0.7, epoch=0):
    model.eval()
    val_loss = 0.0
    all_preds, all_labels = [], []
    with torch.no_grad():
        for batch_idx, batch in enumerate(dataloader):
            images = batch["image"].to(device)
            labels = batch["label"].to(device)
            outputs, clip_features = model(images)
            if batch_idx == 0:
                features = model.feature_map_layer3
                x = model.decoder(features)
                attention_map = model.attention(x)
                mean = torch.tensor([0.485, 0.456, 0.406]).view(3, 1, 1)
                std = torch.tensor([0.229, 0.224, 0.225]).view(3, 1, 1)
                denorm_images = images.cpu() * std + mean
                plt.figure(figsize=(10, 5))
                for i in range(min(4, images.size(0))):
                    plt.subplot(2, 4, i+1)
                    img = denorm_images[i].permute(1, 2, 0).numpy()
                    img = np.clip(img, 0, 1)
                    plt.imshow(img)
                    plt.title("Input Image")
                    plt.axis('off')
                    plt.subplot(2, 4, i+5)
                    plt.imshow(attention_map[i, 0].cpu().numpy(), cmap='viridis')
                    plt.title("Attention Map")
                    plt.axis('off')
                plt.savefig(f'attention_maps_epoch_{epoch+1}.png')
                plt.close()
            ce_loss = criterion(outputs, labels)
            contrastive_loss = model.compute_contrastive_loss(clip_features, text_embeddings, labels, temperature)
            loss = alpha * ce_loss + (1 - alpha) * contrastive_loss
            val_loss += loss.item()
            preds = torch.argmax(outputs, dim=1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
    avg_loss = val_loss / len(dataloader)
    acc = accuracy_score(all_labels, all_preds)
    print("Validation Confusion Matrix:")
    print(confusion_matrix(all_labels, all_preds))
    print("Validation Classification Report:")
    print(classification_report(all_labels, all_preds, target_names=categories))
    return avg_loss, acc, all_preds, all_labels

# =============================
# TRAINING LOOP
# =============================
def main():
    train_losses, val_losses, train_accs, val_accs = [], [], [], []
    best_val_loss = float('inf')
    counter = 0
    best_epoch = 0
    best_model_path = "best_fcn_clip_model.pth"

    class_names = categories

    # Model initialization
    clip_model = FLAIRModel(
        vision_type='resnet_v1',
        from_checkpoint=True,
        weights_path=None,
        image_size=512,
        caption="Fundus image showing [CLS]",
        projection=True,
        norm_features=True
    ).to(device)

    model = FCN_CLIP(clip_model, num_classes=len(categories)).to(device)

    # Handle checkpoint loading with potential mismatch
    if os.path.exists(best_model_path):
        checkpoint = torch.load(best_model_path)
        model_dict = model.state_dict()
        checkpoint_dict = {k: v for k, v in checkpoint.items() if k in model_dict and v.shape == model_dict[k].shape}
        model_dict.update(checkpoint_dict)
        model.load_state_dict(model_dict, strict=False)
        print("‚úÖ Partially loaded best_fcn_clip_model.pth with compatible weights, reinitializing mismatched layers")
        # Reinitialize spatial_gate weights due to kernel size change
        nn.init.kaiming_normal_(model.attention.spatial_gate[0].weight, mode='fan_out', nonlinearity='relu')
        if model.attention.spatial_gate[0].bias is not None:
            nn.init.constant_(model.attention.spatial_gate[0].bias, 0)
        model.use_fusion = False
    else:
        print("No checkpoint found, enabling layer2 fusion")
        model.use_fusion = True
        nn.init.kaiming_normal_(model.attention.spatial_gate[0].weight, mode='fan_out', nonlinearity='relu')
        if model.attention.spatial_gate[0].bias is not None:
            nn.init.constant_(model.attention.spatial_gate[0].bias, 0)

    text_embed_dict, text_embeds = clip_model.compute_text_embeddings(categories, domain_knowledge=True)
    text_embeds = text_embeds.to(device)

    criterion = FocalLoss(alpha=class_weights.to(device), gamma=2.0)
    optimizer = torch.optim.AdamW(model.parameters(), lr=2e-4, weight_decay=1e-3)
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=2)

    for epoch in range(EPOCHS):
        torch.cuda.empty_cache()
        train_loss, train_acc = train_hybrid_epoch(model, loaders["train"], text_embeds, optimizer, criterion, temperature, alpha)
        train_losses.append(train_loss)
        train_accs.append(train_acc)
        val_loss, val_acc, _, _ = validate(model, loaders["val"], text_embeds, criterion, temperature, alpha, epoch)
        val_losses.append(val_loss)
        val_accs.append(val_acc)
        print(f"üìö Epoch {epoch+1}/{EPOCHS} - Train Loss: {train_loss:.4f} - Train Acc: {train_acc:.4f}")
        print(f"Validation Loss: {val_loss:.4f} - Validation Acc: {val_acc:.4f}")

        scheduler.step(val_loss)

        if val_loss < best_val_loss:
            best_val_loss = val_loss
            counter = 0
            best_epoch = epoch + 1
            torch.save(model.state_dict(), best_model_path)
            print(f"üíæ Best model saved at epoch {epoch+1} to {best_model_path}")
        else:
            counter += 1

        if counter >= patience:
            print("üõë Early stopping triggered")
            break

    # Í∏∞Ï°¥ ÌïôÏäµ Í≥°ÏÑ† Í∑∏ÎûòÌîÑ Ï†ÄÏû•
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(train_losses, label="Train Loss")
    plt.plot(val_losses, label="Val Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title("Loss Curve")
    plt.legend()
    plt.grid(True)
    plt.subplot(1, 2, 2)
    plt.plot(train_accs, label="Train Acc")
    plt.plot(val_accs, label="Val Acc")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.title("Accuracy Curve")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig('learning_curves.png')
    plt.close()

    # ÌïôÏäµ Ï¢ÖÎ£å ÌõÑ train/val lossÏôÄ accuracy Í∑∏ÎûòÌîÑ ÌëúÏãú
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(train_losses, label="Train Loss", marker='o')
    plt.plot(val_losses, label="Validation Loss", marker='o')
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title("Train and Validation Loss Over Epochs")
    plt.legend()
    plt.grid(True)
    plt.subplot(1, 2, 2)
    plt.plot(train_accs, label="Train Accuracy", marker='o')
    plt.plot(val_accs, label="Validation Accuracy", marker='o')
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.title("Train and Validation Accuracy Over Epochs")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    print("\nüîç Evaluating on Test Set")
    model_file = best_model_path
    if not os.path.exists(model_file):
        raise FileNotFoundError(f"Best model file {model_file} not found")

    try:
        model.load_state_dict(torch.load(model_file))
    except Exception as e:
        print(f"‚ùå Failed to load model: {e}")
        return None, None

    model.eval()
    all_preds, all_labels, all_probs = [], [], []
    with torch.no_grad():
        for batch_idx, batch in enumerate(tqdm(loaders["test"], desc="Testing")):
            try:
                torch.cuda.empty_cache()
                images = batch["image"].to(device)
                labels = batch["label"].to(device)
                outputs, _ = model(images)
                probs = F.softmax(outputs, dim=1)
                preds = torch.argmax(probs, dim=1)
                all_preds.extend(preds.cpu().numpy())
                all_labels.extend(labels.cpu().numpy())
                all_probs.extend(probs.cpu().numpy())
                print(f"[DEBUG] Processed batch {batch_idx+1}/{len(loaders['test'])}")
            except Exception as e:
                print(f"‚ùå Error in batch {batch_idx+1}: {e}")
                logging.error(f"Test batch {batch_idx+1} error: {e}")
                continue

    if len(all_preds) == 0:
        print("‚ùå No valid predictions made, evaluation failed")
        return None, None

    accuracy = accuracy_score(all_labels, all_preds)
    print(f"\nTest Accuracy: {accuracy:.4f} ({accuracy*100:.2f}%)")
    print("\nConfusion Matrix:")
    print(confusion_matrix(all_labels, all_preds))
    print("\nClassification Report:")
    print(classification_report(all_labels, all_preds, target_names=class_names))
    onehot_labels = F.one_hot(torch.tensor(all_labels), num_classes=len(class_names)).numpy()
    auroc = roc_auc_score(onehot_labels, np.array(all_probs), multi_class='ovr')
    print(f"\nAUROC (OvR): {auroc:.4f}")

    return np.array(all_labels), np.array(all_probs)

# =============================
# VISUALIZATION
# =============================
def compute_optimal_thresholds(model, dataloader, num_classes):
    model.eval()
    all_probs, all_labels = [], []
    with torch.no_grad():
        for batch in dataloader:
            images = batch["image"].to(device)
            labels = batch["label"].to(device)
            outputs, _ = model(images)
            probs = F.softmax(outputs, dim=1)
            all_probs.extend(probs.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    all_probs = np.array(all_probs)
    all_labels = np.array(all_labels)
    y_true_bin = label_binarize(all_labels, classes=list(range(num_classes)))

    optimal_thresholds = []
    for i in range(num_classes):
        fpr, tpr, thresholds = roc_curve(y_true_bin[:, i], all_probs[:, i])
        j_scores = tpr - fpr
        best_thresh = thresholds[np.argmax(j_scores)]
        optimal_thresholds.append(max(0.1, best_thresh))

    return optimal_thresholds

def visualize_results(all_labels, all_probs, class_names, save_dir="./results"):
    os.makedirs(save_dir, exist_ok=True)
    all_labels = np.array(all_labels)
    all_probs = np.array(all_probs)

    if all_labels.shape[0] != all_probs.shape[0]:
        raise ValueError(f"all_labelsÏôÄ all_probsÏùò ÏÉòÌîå ÏàòÍ∞Ä ÏùºÏπòÌïòÏßÄ ÏïäÏäµÎãàÎã§: {all_labels.shape[0]} vs {all_probs.shape[0]}")
    if all_probs.shape[1] != num_classes:
        raise ValueError(f"all_probsÏùò ÌÅ¥ÎûòÏä§ ÏàòÍ∞Ä {num_classes}ÏôÄ ÏùºÏπòÌïòÏßÄ ÏïäÏäµÎãàÎã§: {all_probs.shape[1]}")
    if not np.all(all_probs >= 0) or not np.all(all_probs <= 1):
        raise ValueError("all_probsÎäî 0Í≥º 1 ÏÇ¨Ïù¥Ïùò ÌôïÎ•†Ïù¥Ïñ¥Ïïº Ìï©ÎãàÎã§.")
    if not np.all(np.isin(all_labels, range(num_classes))):
        raise ValueError(f"all_labelsÎäî 0ÏóêÏÑú {num_classes-1} ÏÇ¨Ïù¥Ïùò Ï†ïÏàòÏù¥Ïñ¥Ïïº Ìï©ÎãàÎã§.")
    if all_labels.shape[0] != 812:
        print(f"‚ö†Ô∏è ÏÉòÌîå ÏàòÍ∞Ä ÏòàÏÉÅ(812)Í≥º Îã§Î¶ÖÎãàÎã§: {all_labels.shape[0]}")

    # Compute optimal thresholds
    optimal_thresholds = compute_optimal_thresholds(model, loaders["val"], num_classes)
    print(f"Optimal Thresholds: {optimal_thresholds}")

    y_true = all_labels
    y_score = all_probs
    y_true_bin = label_binarize(y_true, classes=list(range(num_classes)))

    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    for i in range(num_classes):
        fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], y_score[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])
        print(f"Class {class_names[i]}: Optimal Threshold = {optimal_thresholds[i]:.3f}, AUC = {roc_auc[i]:.3f}")

    plt.figure(figsize=(10, 8))
    for i in range(num_classes):
        plt.plot(fpr[i], tpr[i], label=f"{class_names[i]} (AUC = {roc_auc[i]:.2f})")
    plt.plot([0, 1], [0, 1], "k--", label="Random")
    plt.title("AUROC Curve per Class")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(os.path.join(save_dir, 'roc_curve.png'))
    plt.show()

    threshold_preds = []
    for probs in y_score:
        pred_scores = [probs[i] for i in range(num_classes)]
        max_prob = max(pred_scores)
        pred = [int(probs[i] >= optimal_thresholds[i] and probs[i] == max_prob) for i in range(num_classes)]
        if sum(pred) == 0:
            pred = [int(i == np.argmax(pred_scores))]
        threshold_preds.append(np.argmax(pred))

    print("\nüìù Classification Report (Optimal Threshold):")
    print(classification_report(y_true, threshold_preds, target_names=class_names))

    cm = confusion_matrix(y_true, threshold_preds)
    cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis] * 100
    cm_normalized = np.nan_to_num(cm_normalized, nan=0.0)

    annot_labels = []
    for i in range(cm.shape[0]):
        row_labels = []
        for j in range(cm.shape[1]):
            row_labels.append(f"{cm[i,j]}\n({cm_normalized[i,j]:.1f}%)")
        annot_labels.append(row_labels)

    plt.figure(figsize=(10, 8))
    ax = sns.heatmap(cm_normalized,
                     annot=np.array(annot_labels),
                     fmt="",
                     cmap="Blues",
                     vmin=0,
                     vmax=100,
                     xticklabels=class_names,
                     yticklabels=class_names,
                     cbar_kws={'label': 'Recall (%)'})
    plt.title("Confusion Matrix\n(Raw Counts with Row-wise %)", pad=20)
    plt.xlabel("Predicted Label", labelpad=15)
    plt.ylabel("True Label", labelpad=15)
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    for text in ax.texts:
        parts = text.get_text().split('\n')
        percent = float(parts[1].strip('()%'))
        text.set_color("white" if percent > 50 else "black")
    plt.tight_layout()
    plt.savefig(os.path.join(save_dir, 'confusion_matrix.png'))
    plt.show()  # Ensure heatmap is displayed

if __name__ == "__main__":
    all_labels, all_probs = main()
    if all_labels is not None and all_probs is not None:
        visualize_results(all_labels, all_probs, categories, save_dir="./results")
```

```python
%matplotlib inline
import os
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
import matplotlib.pyplot as plt
from torchvision import transforms
from PIL import Image
from model import FLAIRModel
from scipy.ndimage import gaussian_filter
import logging
from skimage.metrics import structural_similarity as ssim

# =============================
# LOGGING SETUP
# =============================
logging.basicConfig(filename='heatmap_errors.log', level=logging.ERROR,
                    format='%(asctime)s - %(levelname)s - %(message)s')

# =============================
# CBAM ATTENTION MODULE
# =============================
class CBAM(nn.Module):
    def __init__(self, channels, reduction=16):
        super(CBAM, self).__init__()
        self.channel_gate = nn.Sequential(
            nn.AdaptiveAvgPool2d(1),
            nn.Conv2d(channels, channels // reduction, 1),
            nn.ReLU(),
            nn.Conv2d(channels // reduction, channels, 1),
            nn.Sigmoid()
        )
        self.spatial_gate = nn.Sequential(
            nn.Conv2d(channels, 1, 7, padding=3),
            nn.Sigmoid()
        )

    def forward(self, x):
        channel_attention = self.channel_gate(x)
        x = x * channel_attention
        spatial_attention = self.spatial_gate(x)
        x = x * spatial_attention
        return x

# =============================
# FCN-CLIP MODEL
# =============================
class FCN_CLIP(nn.Module):
    def __init__(self, clip_model, num_classes=5, dropout_rate=0.3):
        super(FCN_CLIP, self).__init__()
        self.clip_model = clip_model
        self.vision_model = clip_model.vision_model
        self.feature_map_layer3 = None
        self.feature_map_layer2 = None
        if hasattr(self.vision_model.model, 'layer3'):
            self.vision_model.model.layer3.register_forward_hook(self.get_layer3_features)
        if hasattr(self.vision_model.model, 'layer2'):
            self.vision_model.model.layer2.register_forward_hook(self.get_layer2_features)

        self.use_fusion = False
        self.fusion = nn.Conv2d(1024 + 512, 1024, 1)

        self.decoder = nn.Sequential(
            nn.Conv2d(1024, 512, 3, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU(),
            nn.Dropout2d(dropout_rate),
            nn.Upsample(scale_factor=2, mode='bilinear'),
            nn.Conv2d(512, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU()
        )

        self.attention = CBAM(128)

        self.classifier = nn.Sequential(
            nn.AdaptiveAvgPool2d((1, 1)),
            nn.Flatten(),
            nn.Linear(128, num_classes)
        )

    def get_layer3_features(self, module, input, output):
        self.feature_map_layer3 = output

    def get_layer2_features(self, module, input, output):
        self.feature_map_layer2 = output

    def forward(self, x):
        with torch.no_grad():
            clip_features = self.vision_model(x)
        if self.feature_map_layer3 is not None:
            if self.use_fusion and self.feature_map_layer2 is not None:
                layer3 = F.interpolate(self.feature_map_layer3, size=self.feature_map_layer2.shape[2:], mode='bilinear')
                features = torch.cat([layer3, self.feature_map_layer2], dim=1)
                features = self.fusion(features)
            else:
                features = self.feature_map_layer3
        else:
            features = clip_features.view(clip_features.size(0), 1024, 1, 1)
            features = F.interpolate(features, scale_factor=32, mode='nearest')
        x = self.decoder(features)
        attention_map = self.attention(x)
        x = x * attention_map
        outputs = self.classifier(x)
        return outputs, clip_features, features, attention_map

# =============================
# CONFIG
# =============================
device = 'cuda' if torch.cuda.is_available() else 'cpu'
categories = [
    "no diabetic retinopathy",
    "mild diabetic retinopathy",
    "moderate diabetic retinopathy",
    "severe diabetic retinopathy",
    "proliferative diabetic retinopathy"
]
num_classes = len(categories)
model_checkpoint = "best_fcn_clip_model.pth"
sample_paths = [
    "./train_images/0024cdab0c1e.png",  # Class 1 (Mild)
    "./train_images/000c1434d8d7.png"   # Class 2 (Moderate)
]
sample_labels = [1, 2]  # Mild, Moderate

# Î≥ëÎ≥Ä Ï£ºÏÑù ÎßàÏä§ÌÅ¨ Í≤ΩÎ°ú (ÌïÑÏöî Ïãú ÏÇ¨Ïö©, ÏóÜÏúºÎ©¥ None Ïú†ÏßÄ)
lesion_mask_paths = [
    None,  # "./masks/0024cdab0c1e_mask.png",
    None   # "./masks/000c1434d8d7_mask.png"
]

# ÌÅ¥ÎûòÏä§Î≥ÑÎ°ú ÏÑ†Î≥ÑÎêú CLIP Ï±ÑÎÑê (ÏòàÏãú, Ïã§Ï†úÎ°úÎäî Î∂ÑÏÑù ÌïÑÏöî)
class_specific_indices = {
    1: [10, 20, 30, 40, 50],  # MildÏóê Î∞òÏùëÌïòÎäî Ï±ÑÎÑê (ÏûÑÏùò)
    2: [15, 25, 35, 45, 55]   # ModerateÏóê Î∞òÏùëÌïòÎäî Ï±ÑÎÑê (ÏûÑÏùò)
}

# =============================
# DATA AUGMENTATION
# =============================
val_test_transform = transforms.Compose([
    transforms.Resize((512, 512)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# =============================
# VISUALIZATION FUNCTION
# =============================
def calculate_iou(heatmap, gt_mask, threshold=0.5):
    heatmap_bin = (heatmap > threshold).astype(np.uint8)
    gt_bin = (gt_mask > 0).astype(np.uint8)
    intersection = np.logical_and(heatmap_bin, gt_bin).sum()
    union = np.logical_or(heatmap_bin, gt_bin).sum()
    iou = intersection / (union + 1e-6)
    return iou

def visualize_attention(model, sample_paths, sample_labels, class_names, lesion_mask_paths=None, save_dir="./attention_maps"):
    os.makedirs(save_dir, exist_ok=True)
    model.eval()

    # Initialize figure for comparison
    fig, axes = plt.subplots(len(sample_paths), 3, figsize=(15, 5 * len(sample_paths)))

    for i, (path, label, mask_path) in enumerate(zip(sample_paths, sample_labels, lesion_mask_paths if lesion_mask_paths else [None]*len(sample_paths))):
        if not os.path.exists(path):
            logging.error(f"Image not found: {path}")
            print(f"‚ùå Image not found: {path}")
            continue
        try:
            original_image = Image.open(path).convert('RGB')
            original_image = original_image.resize((512, 512))
        except Exception as e:
            logging.error(f"Failed to load image {path}: {e}")
            print(f"‚ùå Failed to load image {path}: {e}")
            continue

        # Load lesion mask if available
        gt_mask = None
        if mask_path and os.path.exists(mask_path):
            try:
                gt_mask = np.array(Image.open(mask_path).resize((512, 512)))
            except Exception as e:
                logging.error(f"Failed to load lesion mask {mask_path}: {e}")
                print(f"‚ùå Failed to load lesion mask {mask_path}: {e}")

        # Prepare image tensor
        image = val_test_transform(original_image).unsqueeze(0).to(device)

        # Forward pass
        try:
            with torch.no_grad():
                outputs, clip_features, clip_feature_map, attention_map = model(image)
        except Exception as e:
            logging.error(f"Forward pass failed for {path}: {e}")
            print(f"‚ùå Forward pass failed for {path}: {e}")
            continue

        # CLIP Heatmap with class-specific channels
        try:
            if label in class_specific_indices:
                indices = torch.tensor(class_specific_indices[label]).to(device)
                clip_heatmap = torch.mean(clip_feature_map[:, indices], dim=1).squeeze().cpu().numpy()
            else:
                top_k = int(clip_feature_map.size(1) * 0.1)
                values, indices = torch.topk(torch.mean(clip_feature_map, dim=(2, 3)), top_k, dim=1)
                clip_heatmap = torch.mean(clip_feature_map[:, indices[0]], dim=1).squeeze().cpu().numpy()
            clip_heatmap = gaussian_filter(clip_heatmap, sigma=2.0)
            clip_heatmap = np.maximum(clip_heatmap, 0)
            clip_heatmap /= clip_heatmap.max() + 1e-6
            clip_heatmap = F.interpolate(
                torch.from_numpy(clip_heatmap).unsqueeze(0).unsqueeze(0).to(device),
                size=(512, 512),
                mode='bilinear',
                align_corners=False
            ).squeeze().cpu().numpy()
            print(f"CLIP Heatmap {i+1} ({class_names[label]}) - Min: {clip_heatmap.min():.4f}, Max: {clip_heatmap.max():.4f}, Mean: {clip_heatmap.mean():.4f}")
        except Exception as e:
            logging.error(f"CLIP heatmap generation failed for {path}: {e}")
            print(f"‚ùå CLIP heatmap generation failed for {path}: {e}")
            continue

        # CBAM Heatmap
        try:
            cbam_heatmap = torch.mean(attention_map, dim=1).squeeze().cpu().numpy()
            cbam_heatmap = gaussian_filter(cbam_heatmap, sigma=2.0)
            cbam_heatmap = np.maximum(cbam_heatmap, 0)
            cbam_heatmap /= cbam_heatmap.max() + 1e-6
            cbam_heatmap = F.interpolate(
                torch.from_numpy(cbam_heatmap).unsqueeze(0).unsqueeze(0).to(device),
                size=(512, 512),
                mode='bilinear',
                align_corners=False
            ).squeeze().cpu().numpy()
            print(f"CBAM Heatmap {i+1} ({class_names[label]}) - Min: {cbam_heatmap.min():.4f}, Max: {cbam_heatmap.max():.4f}, Mean: {cbam_heatmap.mean():.4f}")
        except Exception as e:
            logging.error(f"CBAM heatmap generation failed for {path}: {e}")
            print(f"‚ùå CBAM heatmap generation failed for {path}: {e}")
            continue

        # Calculate SSIM
        try:
            similarity = ssim(clip_heatmap, cbam_heatmap, data_range=1.0)
            print(f"SSIM between CLIP and CBAM heatmap {i+1}: {similarity:.4f}")
        except Exception as e:
            logging.error(f"SSIM calculation failed for {path}: {e}")
            print(f"‚ùå SSIM calculation failed for {path}: {e}")

        # Calculate IoU with ground truth if available
        if gt_mask is not None:
            try:
                clip_iou = calculate_iou(clip_heatmap, gt_mask)
                cbam_iou = calculate_iou(cbam_heatmap, gt_mask)
                print(f"CLIP IoU with lesion mask {i+1}: {clip_iou:.4f}")
                print(f"CBAM IoU with lesion mask {i+1}: {cbam_iou:.4f}")
            except Exception as e:
                logging.error(f"IoU calculation failed for {path}: {e}")
                print(f"‚ùå IoU calculation failed for {path}: {e}")

        # Denormalize original image
        mean = torch.tensor([0.485, 0.456, 0.406]).view(3, 1, 1)
        std = torch.tensor([0.229, 0.224, 0.225]).view(3, 1, 1)
        denorm_image = image.cpu() * std + mean
        original_array = denorm_image.squeeze().permute(1, 2, 0).numpy()
        original_array = np.clip(original_array, 0, 1)

        # Plotting
        axes[i, 0].imshow(original_array)
        axes[i, 0].set_title(f"Original\nClass: {class_names[label]}")
        axes[i, 0].axis('off')

        axes[i, 1].imshow(original_array)
        axes[i, 1].imshow(clip_heatmap, cmap='jet', alpha=0.4)
        axes[i, 1].set_title("CLIP Attention Map")
        axes[i, 1].axis('off')

        axes[i, 2].imshow(original_array)
        axes[i, 2].imshow(cbam_heatmap, cmap='jet', alpha=0.6)
        axes[i, 2].set_title("CBAM Attention Map")
        axes[i, 2].axis('off')

        # Save individual heatmaps
        plt.figure(figsize=(5, 5))
        plt.imshow(original_array)
        plt.imshow(clip_heatmap, cmap='jet', alpha=0.4)
        plt.title(f"CLIP Heatmap - {class_names[label]}")
        plt.axis('off')
        clip_path = os.path.join(save_dir, f"clip_heatmap_{i+1}_{class_names[label].replace(' ', '_')}.png")
        plt.savefig(clip_path)
        plt.close()
        print(f"‚úÖ Saved CLIP heatmap to {clip_path}")

        plt.figure(figsize=(5, 5))
        plt.imshow(original_array)
        plt.imshow(cbam_heatmap, cmap='jet', alpha=0.6)
        plt.title(f"CBAM Heatmap - {class_names[label]}")
        plt.axis('off')
        cbam_path = os.path.join(save_dir, f"cbam_heatmap_{i+1}_{class_names[label].replace(' ', '_')}.png")
        plt.savefig(cbam_path)
        plt.close()
        print(f"‚úÖ Saved CBAM heatmap to {cbam_path}")

    # Save and display comparison plot
    plt.tight_layout()
    comparison_path = os.path.join(save_dir, "attention_comparison.png")
    plt.savefig(comparison_path)
    plt.show()
    print(f"‚úÖ Saved attention_comparison.png to {comparison_path}")

    return comparison_path

# =============================
# MAIN
# =============================
def main():
    torch.cuda.empty_cache()

    # Initialize CLIP model
    try:
        clip_model = FLAIRModel(
            vision_type='resnet_v1',
            from_checkpoint=True,
            weights_path=None,
            image_size=512,
            caption="Fundus image showing [CLS]",
            projection=True,
            norm_features=True
        ).to(device)
    except Exception as e:
        logging.error(f"Failed to initialize FLAIRModel: {e}")
        print(f"‚ùå Failed to initialize FLAIRModel: {e}")
        return

    # Initialize FCN_CLIP model
    model = FCN_CLIP(clip_model, num_classes=num_classes).to(device)
    if os.path.exists(model_checkpoint):
        try:
            model.load_state_dict(torch.load(model_checkpoint), strict=True)
            print(f"‚úÖ Loaded {model_checkpoint} successfully")
        except Exception as e:
            logging.error(f"Failed to load checkpoint {model_checkpoint}: {e}")
            print(f"‚ùå Failed to load checkpoint {model_checkpoint}: {e}")
            return
    else:
        logging.error(f"Model checkpoint {model_checkpoint} not found")
        print(f"‚ùå Model checkpoint {model_checkpoint} not found")
        return

    # Generate and save heatmaps
    comparison_path = visualize_attention(model, sample_paths, sample_labels, categories, lesion_mask_paths)
    if comparison_path:
        print(f"‚úÖ Visualization complete. Comparison plot saved at: {comparison_path}")

if __name__ == "__main__":
    main()
```
